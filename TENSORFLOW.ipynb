```python
# creatiing sokem data to vie and fit
import numpy as np
import matplotlib.pyplot as plt

# create the features
X=np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,14.0])
y=np.array([3.0,6.0,9.0,12.0,15.0,18.0,21.0,24.0])

plt.scatter(X,y)
```

```python
input_shape=X[0].shape
output_shape=y[0].shape
input_shape,output_shape
```

```python
# turn out the numpy arrays into tyhe tensors
X=tf.cast(tf.constant(X),dtype=tf.float32)
y=tf.cast(tf.constant(y), dtype=tf.float32)
```

```python
plt.scatter(X,y)
```

```python
# steps in modelling with the Tensorflow
#1) Get the data ready as turn into the tensors
# 2) Build or pick a pretrained model
# 3) fit the model to the data and make a prediction
# 4) evaluate the model
# 5) improve through the experiments
# 6) Save and reload the trained model
```

```python
tf.random.set_seed(42)

model=tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])
# mae ->  means of the absioluet of y true and the y pred
# 2 Compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])

# 3 fit the model
model.fit(tf.expand_dims(X,axis=-1),y,epochs=5)
```

```python
 # check out X and y
X,y
```

```python
# tru and make a prediction usng a model
prediction = model.predict(np.array([[17.0]]))
print(prediction)
```

```python
# Steps in Improving the Model using the Tensorflow
# create the model, compile the model and fit the model.

# create the moel here we might add more layers within each hidden layers change the activation functions of each laer
# here we might change the optimization function or perhaps the lerning rate of teh optimizing function
# here we mught fit the model.
```

```python
model=tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
X = tf.expand_dims(X, axis=-1)
model.fit(X, y, epochs=100)
```

```python
# Larger  Rebuild Model Code
# create the model with tge extra hidden layer
model=tf.keras.Sequential([
  tf.keras.layers.Dense(100,activation=None),
  # tf.keras.layers.Dense(100,"relu"),
  # tf.keras.layers.Dense(100,"relu"),
  tf.keras.layers.Dense(1)
])# compile the model
model.compile(loss="mae",
              optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              metrics=["mae"])
# fit the model
model.fit(X,y,epochs=100)
```

```python
tf.random.set_seed(42)

model=tf.keras.Sequential([
    tf.keras.layers.Dense(10,input_shape=[1],name="input_layer") ,# 1 input for the 1 output
    tf.keras.layers.Dense(1,name="output_layer")
],name="model_1")
model.compile(loss=tf.keras.losses.mae,
              optimizer=tf.keras.optimizers.SGD(),
              metrics=["mae"])
```

```python
model.summary()
```

```python
# total_prams=total number of parameters in the model
# how many different patterns our model is going to learn
# trainable parameters these are parameters the model can update as it trains
# non-trainable params- these parametrs aren't update s during training
```

```python
# lets fit our model
model.fit(X_train,y_train,epochs=100,verbose=0)
```
